```python
import bycycle
```

```python
from bycycle.filt import amp_by_time, phase_by_time, bandpass_filter
from bycycle.sim import sim_noisy_bursty_oscillator
```

```python
f_alpha = (8, 12)
N_seconds_filter = .5
sig = np.squeeze(sig) #squeeze her because 2d into 1
# Compute amplitude and phase
signal_filt = bandpass_filter(sig, fs, f_alpha, N_seconds=N_seconds_filter)
theta_amp = amp_by_time(sig, fs, f_alpha,
                       filter_kwargs={'N_seconds': N_seconds_filter})
theta_phase = phase_by_time(sig, fs, f_alpha,
                            filter_kwargs={'N_seconds': N_seconds_filter})
```

```python
t = np.arange(0, len(sig)/fs, 1/fs)
tlim = (2, 6)
tidx = np.logical_and(t>=tlim[0], t<tlim[1])

plt.figure(figsize=(12,6))
plt.subplot(3,1,1)
plt.plot(t[tidx], sig[tidx], 'k')
plt.xlim(tlim)
plt.ylabel('Voltage (mV)')

plt.subplot(3,1,2)
plt.plot(t[tidx], signal_filt[tidx], 'k', alpha=.5)
plt.plot(t[tidx], theta_amp[tidx], 'r')
plt.xlim(tlim)
plt.ylabel('Oscillation amplitude', color='r')

plt.subplot(3,1,3)
plt.plot(t[tidx], theta_phase[tidx], 'r')
plt.xlim(tlim)
plt.ylabel('Phase (radians)', color='r')
```

```python
from bycycle.filt import lowpass_filter
f_theta = (4, 10)
f_lowpass = 30
N_seconds = .1

signal_low = lowpass_filter(sig, fs, f_lowpass,
                            N_seconds=N_seconds, remove_edge_artifacts=False)


# Plot signal
t = np.arange(0, len(sig)/fs, 1/fs)
tlim = (2, 5)
tidx = np.logical_and(t>=tlim[0], t<tlim[1])

plt.figure(figsize=(12, 2))
plt.plot(t[tidx], sig[tidx], '.5')
plt.plot(t[tidx], signal_low[tidx], 'k')
plt.xlim(tlim)
```

```python
from bycycle.filt import bandpass_filter
from bycycle.cyclepoints import _fzerorise, _fzerofall, find_extrema

# Narrowband filter signal
N_seconds_theta = .75
signal_narrow = bandpass_filter(sig, fs, f_theta,
                                remove_edge_artifacts=False,
                                N_seconds=N_seconds_theta)

# Find rising and falling zerocrossings (narrowband)
zeroriseN = _fzerorise(signal_narrow)
zerofallN = _fzerofall(signal_narrow)

# Find peaks and troughs (this function also does the above)
Ps, Ts = find_extrema(signal_low, fs, f_theta,
                      filter_kwargs={'N_seconds':N_seconds_theta})
```

```python
tlim = (12, 15)
tidx = np.logical_and(t>=tlim[0], t<tlim[1])
tidxPs = Ps[np.logical_and(Ps>tlim[0]*fs, Ps<tlim[1]*fs)]
tidxTs = Ts[np.logical_and(Ts>tlim[0]*fs, Ts<tlim[1]*fs)]

plt.figure(figsize=(12, 2))
plt.plot(t[tidx], signal_low[tidx], 'k')
plt.plot(t[tidxPs], signal_low[tidxPs], 'b.', ms=10)
plt.plot(t[tidxTs], signal_low[tidxTs], 'r.', ms=10)
plt.xlim(tlim)
```
