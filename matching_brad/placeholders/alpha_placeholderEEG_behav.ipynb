```python
import seaborn as sns
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import mne
```

```python
where_files = r'/home/dcellier/RDSS/AlphaStudy_Data/eegData/eeg_behavior_data/'
preproc_direc='/data/backed_up/shared/AlphaStudy_data/placeholders/preproc_EEG_matchingBrad/'
files=os.listdir(where_files)
listDirectory=[f[:3] for f in files if 'matchingBrad' in f ]
```

```python
subs_list=['197','206','211','214','223',
             '224','225','226','228','233',
             '234','237','241','244',
             '247','240','231','251',
             '252']
```

```python
len(subs_list)
```

```python
allSubs=pd.DataFrame()
allDfsList=[]
for sub in subs_list:
    thisSubDf=pd.DataFrame() #compile all of the csv's for ea subject
    preproc_Cues=mne.read_epochs(preproc_direc+sub+'/cue-epo.fif')
    thisSub_data=preproc_Cues.metadata
    thisSub_data['subject']=sub
    allDfsList.append(thisSub_data)
    allSubs=allSubs.append(thisSub_data)
            
```

```python
allSubs[allSubs['block']>4].groupby(['trial_type','SOA']).mean()
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(y='RT',x='trial_type',hue='SOA',order=['dis','neut','tar'],hue_order=['short','long'],
               data=allSubs[allSubs['block']<=4],palette='PuBu')
plt.legend(title='SOA: Long or Short')
#plt.ylim=(0,2)
plt.xlabel("\nTrial Type",size=25)
plt.ylabel("RT",size=25)
plt.title("First Five Blocks",size=25)
```
