```python
import seaborn as sns
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
where_files = r'/home/dcellier/RDSS/AlphaStudy_Data/eegData/eeg_behavior_data/'
files=os.listdir(where_files)
listDirectory=[f[:3] for f in files if 'matchingBrad' in f ]
```

```python
np.unique(listDirectory)
```

```python
subs_list=['193', '196', '197', '198', '204', '206', '211', '212', '213',
       '214']
```

```python
allSubs=pd.DataFrame()
allDfsList=[]
for sub in subs_list:
    thisSubDf=pd.DataFrame() #compile all of the csv's for ea subject
    for file in os.listdir(where_files):
        if (sub in file) and ("TRUNCATED" not in file):
            print(file)
            thisSub_data=pd.read_csv(where_files+file) #import subject data if filename matches pattern
            thisSub_data['subject']=sub
            thisSubDf=thisSubDf.append(thisSub_data)
    allDfsList.append(thisSubDf)
    allSubs=allSubs.append(thisSubDf)
            
```

```python
allSubs
```

```python
allSubs.groupby(['trial_type','SOA']).mean()
```

```python
plt.figure(figsize=(8,6))
ax=sns.barplot(y='RT',x='trial_type',hue='SOA',data=allSubs,palette='PuRd')
plt.legend(title='SOA: Long or Short')
#plt.ylim=(0,2)
plt.xlabel("\nTrial Type",size=25)
plt.ylabel("RT",size=25)
```
