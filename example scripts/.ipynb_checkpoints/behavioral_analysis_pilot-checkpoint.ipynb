```python
all_results=pd.DataFrame()
sub_results=[]
#print(all_results)
files=os.listdir(where_files)
for sub in subjects:
    pattern='%s_alpha_pilot_01_2018_Oct_*_*.csv' %sub
    for f in files:
        if fnmatch.fnmatch(f,pattern):
            thisSub=f
    thisSubf=where_files+thisSub
    thisSub_result=pd.read_csv(thisSubf)
    #print(thisSub_result)
    sub_results.append(thisSub_result)
    all_results=all_results.append(thisSub_result,ignore_index=True)
    
```
